```python
milan_data = all_daily_sales[all_daily_sales["tab negozio"] == "ARESE"]
milan_data
```

```python
rome_data = all_daily_sales[(all_daily_sales["tab negozio"] == "RMNAZ") | (all_daily_sales["tab negozio"] == "TIBUR")
                            | (all_daily_sales["tab negozio"] == "TUSC1") | (all_daily_sales["tab negozio"] == "TUSC2")
                            ]
rome_data
```

```python
# Drop unnamed column and listino column, we won't be using those
rome_data = rome_data.drop(["Unnamed: 0", "tab negozio"], axis=1)

# Build datetime index
rome_data["time"] = [dateindex_from_timestamp(timestamp[:10]) for timestamp in rome_data["data"].values]
rome_data = rome_data.set_index(["time"])
rome_data.index = pd.to_datetime(rome_data.index) # make ts

# Sort the data based on the new time index
rome_data.sort_index(inplace=True)

rome_data = rome_data.drop(["data", "codice esterno"], axis=1)

# Aggregate the data to get the total sales of each week
rome_data_daily = rome_data.groupby(pd.Grouper(freq = "D")).agg({
    "qty": np.sum
})

rome_data_daily.to_csv("../datasets/rome_daily.csv")
```
