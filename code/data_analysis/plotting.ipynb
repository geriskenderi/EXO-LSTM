```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.io as pltio
```

```python
# Daily and weekly data
milan_data_df = pd.read_csv("../../datasets/ts_data/daily_aug/aug_milan_cc_arese_data.csv", index_col=[0], parse_dates=True)
milan_data_df_w = pd.read_csv("../../datasets/ts_data/weekly_aug/aug_milan_cc_arese_data_w.csv", index_col=[0], parse_dates=True)

turin_data_df = pd.read_csv("../../datasets/ts_data/daily_aug/aug_turin_cc_torin_data.csv", index_col=[0], parse_dates=True)
turin_data_df_w = pd.read_csv("../../datasets/ts_data/weekly_aug/aug_turin_cc_torin_data_w.csv", index_col=[0], parse_dates=True)

rome_data_df = pd.read_csv("../../datasets/ts_data/daily_aug/aug_rome_cc_eur2_data.csv", index_col=[0], parse_dates=True)
rome_data_df_w = pd.read_csv("../../datasets/ts_data/weekly_aug/aug_rome_cc_eur2_data_w.csv", index_col=[0], parse_dates=True)
```

```python
data_d = rome_data_df.copy(deep=True)
data_w = rome_data_df_w.copy(deep=True)
```

```python
plt.figure(figsize=(14,8))
plt.plot(data_d.qty, label="Rome - Daily sales data")
plt.legend()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig("daily_plot", transparent = True)
plt.show()
plt.figure(figsize=(14,8))
plt.plot(data_w.qty, label="Rome - Weekly sales data")
plt.legend()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig("weekly_plot", transparent = True)
plt.show()
```

```python
values = data_d.values
cols_to_plot = [5,8,9]
i=1
plt.figure(figsize=(10,8))
for col in cols_to_plot:
    plt.subplot(len(cols_to_plot), 1, i)
    plt.plot(data_d.iloc[:, col], label=data_d.columns[col])
    plt.legend()
    i += 1
plt.savefig("some_weather_data.png", transparent = True)
plt.tight_layout() 
plt.show()
```
